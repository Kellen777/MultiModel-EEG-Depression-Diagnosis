```python
# 创建文件夹
folder_path1 = '../data/EEG_dataset'
folder_path2 = '../data/EEG_dataset/MDD'
folder_path3 = '../data/EEG_dataset/Control'
if not os.path.exists(folder_path1):
    os.makedirs(folder_path1)
if not os.path.exists(folder_path2):
    os.makedirs(folder_path2)
if not os.path.exists(folder_path3):
    os.makedirs(folder_path3)
```

```python
# MDD类别数据集：将原始mat文件转换为numpy数组
datapath="../data/EEG_128channels_resting_lanzhou_2015"
save_path='../data/EEG_dataset/MDD'
lstpath=os.listdir(datapath)

patren='0201' # 选择属于MDD类别的文件模式，文件名带‘0201’前缀
mdd_list=list(filter(lambda x: patren in x,lstpath))  # 筛选出符合MDD类别模式的文件

for mdd in mdd_list: 
    filepath=os.path.join(datapath,mdd)  # 拼接文件路径
    print(filepath)  # 输出当前文件路径
    datafile=sio.loadmat(filepath)  # 读取mat文件
    
    r = re.compile(".*a0.*")  # 使用正则表达式筛选出包含EEG数据的数组，mat文件中包含不同的模式
    dataff=list(filter(r.match, datafile))  # 根据正则表达式筛选符合条件的EEG数组
    dataff1=dataff[0]
    dataeeg=datafile[dataff1]  # 从mat文件中提取出EEG数据数组
    print(dataeeg.shape)
    
    np.save(os.path.join(save_path,dataff1[1:9]+".npy"),dataeeg)  # 保存EEG数据为.npy格式文件到指定的路径
```

```python
# Control类别数据集：将原始mat文件转换为numpy数组
datapath="../data/EEG_128channels_resting_lanzhou_2015"
save_path='../data/EEG_dataset/Control'
lstpath=os.listdir(datapath)  # 获取指定目录下的所有文件和子文件夹列表

pattern2='0202'  # 选择健康对照（HC）对象的模式1'0202'
control_lst=list(filter(lambda x: pattern2 in x,lstpath))
pattern3='0203'  # 选择健康对照（HC）对象的模式2'0203'
control_lst1=list(filter(lambda x: pattern3 in x,lstpath)) 

contotal=control_lst + control_lst1  # 合并符合两个模式的文件列表

# 遍历所有符合条件的文件，处理方式和前面一致
for cntrol in contotal:
    pathc=os.path.join(datapath,cntrol)  
    print(pathc)  
    datafilec=sio.loadmat(pathc) 
    
    rc = re.compile(".*a0.*") 
    dataffc2=list(filter(rc.match, datafilec))  
    dataffc22=dataffc2[0] 
    dataeeg_c=datafilec[dataffc22]  
    print(dataeeg_c.shape)  
    
    np.save(os.path.join(save_path, dataffc22[1:9] + ".npy"), dataeeg_c) 
```
